```python
def conversation_by_date(file,date):
    
    fhand = open(file,encoding = 'utf8')
    for i in fhand:
        if int(date[3:5])<10 and int(date[:2])<10:   #main problem is date format dd/mm/yyyy whereas in chat it is mm/dd/yy
            if i.startswith(f'{date[4:5]}/{date[1:2]}/{date[8:]}'):
                print(i)
        elif int(date[3:5])>=10 and int(date[:2])<10:    #when date or month is less than 10 it gives 1 instead of 01 so we have to take various where date is single digit
            if i.startswith(f'{date[3:5]}/{date[1:2]}/{date[8:]}'):
                print(i)
        elif int(date[3:5])<10 and int(date[:2])>=10:
            if i.startswith(f'{date[4:5]}/{date[:2]}/{date[8:]}'):
                print(i)
        elif int(date[3:5])>=10 and int(date[:2])>=10:
            if i.startswith(f'{date[3:5]}/{date[:2]}/{date[8:]}'):
                print(i)
```

```python
def converstion_by_person(file,person):   
    fhand = open(file,encoding = 'utf8')
    member_chat={}
    for i in fhand:
        i=i.rstrip() #to remove the '\n' from each line
        try:
            a=i[i.index('-')+2:i.index(':',16,55)] #substringing done to extract name of person messaging 
            b=i[i.index('-')+2:] #extracting chat to that person in that line
        except:
            continue
        member_chat[a] = member_chat.get(a,'')+","+b #adding all chats with name in dictionary seprated by a comma
    return(member_chat[person])
```

```python
def avg_length_by_person(file,person):   
    fhand = open(file,encoding = 'utf8')
    member_chat={}
    member_count={}
    for i in fhand:
        i=i.rstrip() #to remove the '\n' from each line
        try:
            a=i[i.index('-')+2:i.index(':',16,55)]
            b=i[i.index(':',16,55)+2:]
        except:
            continue
        member_chat[a] = member_chat.get(a,0)+len(b) #similarly as in last question a dictionary is craeted but this time for length of string
        member_count[a] = member_count.get(a,0)+1 #new dictionary with count of messages done by single user
    member_avg_count=(member_chat[person]/member_count[person]) #finding avg by dividing len of chat by count of number of msgs tilldate sent by that person
    return(member_avg_count)
```
